```python
import os
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Define the paths to train and validation data
train_path = r"C:\Users\HP\Desktop\Data\Plant Disease Detection\PlantVillage\train"
val_path = r"C:\Users\HP\Desktop\Data\Plant Disease Detection\PlantVillage\val"
```

```python
# Get the list of classes
classes = os.listdir(train_path)

train_counts = []
val_counts = []
```

```python
# Iterate through each class
for class_name in classes:
    # Count images in train set
    train_count = len(os.listdir(os.path.join(train_path, class_name)))
    train_counts.append(train_count)
    
    # Count images in validation set
    val_count = len(os.listdir(os.path.join(val_path, class_name)))
    val_counts.append(val_count)
```

```python
# Create a DataFrame for visualization
import pandas as pd

data = pd.DataFrame({'Class': classes, 'Train Count': train_counts, 'Validation Count': val_counts})
```

```python
# Set up the visualization
plt.figure(figsize=(12, 8))
sns.set(style="whitegrid")
sns.barplot(x="Class", y="Train Count", data=data, color="skyblue", label="Train")
sns.barplot(x="Class", y="Validation Count", data=data, color="salmon", label="Validation")
plt.title('Distribution of Images per Class in Train and Validation Sets')
plt.xlabel('Class')
plt.ylabel('Count of Images')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Define constants for data augmentation
BATCH_SIZE = 16
IMAGE_SIZE = (224, 224)  # Size to resize the images
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
# Data Transformation: Normalization and Rescaling
train_datagen = ImageDataGenerator(rescale=1./255)
val_datagen = ImageDataGenerator(rescale=1./255)
```

```python
# Data Augmentation
train_augmentation = ImageDataGenerator(
    rescale=1./255,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)
```

```python
# Load images from directories and apply transformations
train_generator = train_augmentation.flow_from_directory(
    train_path,
    target_size=IMAGE_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical'
)

val_generator = val_datagen.flow_from_directory(
    val_path,
    target_size=IMAGE_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical'
)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization, Activation
from tensorflow.keras.optimizers import Adam
```

```python
# Define the input shape
height, width, depth = IMAGE_SIZE[0], IMAGE_SIZE[1], 3  # Assuming RGB images
```

```python
# Compile the model
model = Sequential()

# Input layer
model.add(Conv2D(16,(3,3), activation='relu',
                 kernel_initializer='he_uniform', 
                 padding='same', input_shape=(224,224,3))) # filter image by dot product in matrix to find the object in picture
# Hidden layers
model.add(MaxPooling2D(pool_size=(2,2))) # reduce stucture of image even if object slightly move but the result is same  
model.add(Conv2D(32,(3,3), activation='relu',
                 kernel_initializer='he_uniform', 
                 padding='same')) 
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Conv2D(16,(3,3), activation='relu',
                 kernel_initializer='he_uniform', 
                 padding='same')) # filter image by dot product in matrix to find the object in picture
# Hidden layers
model.add(MaxPooling2D(pool_size=(2,2))) # reduce stucture of image even if object slightly move but the result is same
model.add(Conv2D(32,(3,3), activation='relu',
                 kernel_initializer='he_uniform', 
                 padding='same')) # filter image by dot product in matrix to find the object in picture
# Hidden layers
model.add(MaxPooling2D(pool_size=(2,2))) # reduce stucture of image even if object slightly move but the result is same
model.add(Flatten()) # convert all the resultant 2-Dimensional arrays into a single long continuous linear vector
model.add(Dense(64, activation='relu',
                kernel_initializer='he_uniform'))

# avoid overfitting!
model.add(Dropout(0.5))

# Output layer
model.add(Dense(len(classes), activation='softmax'))

# Compile the model         
model.compile(loss="categorical_crossentropy", 
              optimizer=Adam(learning_rate=0.001),
              metrics=['accuracy'])
```

```python
# Print model summary
model.summary()
```

```python
# Save checkpoints during training
checkpoint_path = r"C:\Users\HP\Desktop\Data\Plant Disease Detection\callbacks\cp.ckpt"
checkpoint_dir = os.path.dirname(checkpoint_path)
```

```python
# Create a callback that saves the model's weights
cp_callback = tf.keras.callbacks.ModelCheckpoint(filepath=checkpoint_path,
                                                 save_weights_only=True,
                                                 verbose=1)
```

```python
# training model 
# make sure you truely save checkpoint_path
history = model.fit(train_generator,
                              epochs=5,
                              validation_data=val_generator,
                              callbacks=[cp_callback])
```

```python
# Plot the training history
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Evaluate the model on the validation dataset
val_loss, val_accuracy = model.evaluate(val_generator)

print(f'Validation Loss: {val_loss}')
print(f'Validation Accuracy: {val_accuracy}')
```

```python
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report

# Predict classes for the validation dataset
val_predictions = model.predict(val_generator)
val_predicted_classes = np.argmax(val_predictions, axis=1)
val_true_classes = val_generator.classes

# Get the class labels
class_labels = list(val_generator.class_indices.keys())

# Calculate confusion matrix
conf_matrix = confusion_matrix(val_true_classes, val_predicted_classes)

# Calculate accuracy for each class
class_accuracy = {}
for i in range(len(class_labels)):
    class_accuracy[class_labels[i]] = conf_matrix[i, i] / np.sum(conf_matrix[i, :])

# Print accuracy for each class
print("Accuracy for each class:")
for class_label, accuracy in class_accuracy.items():
    print(f"{class_label}: {accuracy}")

# Print classification report
print("\nClassification Report:")
print(classification_report(val_true_classes, val_predicted_classes, target_names=class_labels))
```
